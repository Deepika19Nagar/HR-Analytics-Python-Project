```python
sns.distplot(hr_data['Age'])
plt.title("Distribution of Age")
plt.show()
```

Most of the 30 to 40 year of age of employees are working as the distributin plot show.

```python
sns.distplot(hr_data['YearsInCurrentRole'])
plt.title("Distribution Years In CurrentRole")
plt.show()
```

This distribution plot show that Majority of employees are in current role for 2-3 years.

```python
sns.distplot(hr_data['TotalWorkingYears'])
plt.title("Distribution of Total Working Years")
plt.show()
```

According to the distribution plot, the total working years lie between 5 to 10 years.

```python
sns.displot(x='YearsAtCompany',data=hr_data,kind='kde',hue="Gender")
plt.title("Distribtion of years at company")
plt.show()
```

According to this distribution plot,we can see that employees have total experience around 5 to 10 years in the company.Male employees have more working experience than the female employees.

```python
hr_data.columns
```

```python
sns.countplot(x="Attrition",data=hr_data,hue='Gender')
plt.title("Count of Attrition")
plt.show()
```

It shows that the attrition rate is minimum but male employees have high attrition rate then the female employees.It means around 700 male and around 500 females are still working in the company and rest of the employees left the company.

```python
sns.barplot(x='Attrition',y='MonthlyIncome',data=hr_data, palette='cividis')
plt.show()
```

One of the main reasons could be the "Monthly Income" that employees have left the company.

```python
sns.barplot(x='Attrition',y='DistanceFromHome',data=hr_data,hue="Gender",estimator='mean', palette=['red', 'skyblue'])
plt.show()
```

If we consider distance from home of the employees, the attrition rate is high in case of female employees. It seems like most of females employees cannot afford daily up down from home to office.

```python
sns.barplot(x="BusinessTravel",y='JobSatisfaction',data=hr_data,hue='Gender',palette=['orange','yellow'])
plt.show()
```

This barplot shows that the female employess travel less than the male employees.

```python
sns.barplot(y="MonthlyIncome",x='JobSatisfaction',data=hr_data,hue="Gender",palette=['purple','pink'])
plt.show()
```

If we see at all the level of job satisfaction in respect to monthly income, the female employees are more satisfied with their jobs other than the male employees. This can also be the reason that why female employees did not left the company other than male employees.

```python
sns.barplot(x="Attrition",y='MonthlyIncome',data=hr_data,hue='MaritalStatus',palette=['red','yellow','green'],estimator='median')
plt.show()
```

This plot shows that most of the employees who left the company were married.

```python
sns.jointplot(x='Age',y='MonthlyIncome',data=hr_data,kind="scatter")
plt.show()
```

Scatter Plot shows that there is linear relationship between Age and Monthly income.
According to data, if the age is more, monthly income also increases.
